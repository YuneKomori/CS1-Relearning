### 課題１

#### 正の整数からなるリストxsを引数にとり、xsの要素の中で３の倍数の個数を返す関数count3(xs)を定義せよ

```python
def count3(xs):
    count=0
    for i in xs:
        if i%3==0:
            count+=1
    print(count)
```

```python
count3([123,456,789,101])
```

### 課題２

#### 単語のリストwordsを引数にとり、空白を挟んでつないだ文字列を返す関数sentence(words)を定義せよ

```python
def sentence(words):
    result=""
    for i in words:
        result+=i
        result+= " "
    print(result)
    
```

```python
sentence(['this','is','iniad'])
```

### 課題 1-1-2

#### 10000円を、金利3%で運用するとします。単利で運用した場合と、複利で運 用した場合について、それぞれの20年間の利子を含む残金を、以下のよう に重ねて描画しなさい。

```python
import matplotlib.pyplot as plt
import math
```

```python
simple_xs= []
simple_ys= []
compound_xs=[]
compound_ys=[]

for x in range(0,21):
    y=10000+300*x
    simple_xs.append(x)
    simple_ys.append(y)

for x in range(0,21):
    y=10000*(1.03)**x
    compound_xs.append(x)
    compound_ys.append(y)


# plot
plt.plot(simple_xs,simple_ys,label='simple')
plt.plot(compound_xs,compound_ys,label='compound')

# タイトルとかをつける
plt.title('simple and compound interest')
plt.xlabel('Year')
plt.ylabel('Amount')
plt.legend()
# 表示
plt.show()
```
