## Matplotlibでグラフを描こう

```python
import matplotlib.pyplot as plt
import math
```

```python
xs= [0,1,2,3]
ys= [0,3,2,3]

plt.plot(xs,ys)
plt.show()
```

```python
xs= [0,1,2,3]
ys= [0,3,2,3]

plt.title('sample graph')
plt.xlabel('x')
plt.ylabel('y')


plt.plot(xs,ys)
plt.show()
```

```python
xs= [0,1,2,3]
ys= [0,3,2,3]
ys2=[3,2,1,0]

plt.plot(xs,ys,label='sample')
plt.plot(xs,ys2,label='sample2')
plt.title('sample graph')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()


plt.plot(xs,ys)
plt.show()
```

### グラフのスタイルの設定

```python
xs= [0,1,2,3]
ys= [0,3,2,3]
ys2=[3,2,1,0]

plt.plot(xs,ys,'o',label='sample')
plt.plot(xs,ys2,'x',label='sample2')

plt.title('sample graph')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()


plt.plot(xs,ys)
plt.show()
```
